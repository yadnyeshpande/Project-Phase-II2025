```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Paths to both image folders
img_folder_1 = "HAM10000/HAM10000_images_part_1"
img_folder_2 = "HAM10000/HAM10000_images_part_2"

# Function to find image in both folders
def find_image(img_id):
    img_path_1 = os.path.join(img_folder_1, img_id + ".jpg")
    img_path_2 = os.path.join(img_folder_2, img_id + ".jpg")
    
    if os.path.exists(img_path_1):
        return img_path_1
    elif os.path.exists(img_path_2):
        return img_path_2
    else:
        return None  # Image not found

# Select 5 random images
sample_images = df.sample(5)  

plt.figure(figsize=(10, 5))
for i, img_id in enumerate(sample_images['image_id']):
    plt.subplot(1, 5, i + 1)
    
    img_path = find_image(img_id)
    if img_path:
        img = mpimg.imread(img_path)
        plt.imshow(img)
        plt.axis("off")
        plt.title(sample_images.iloc[i]['dx'])
    else:
        print(f" Image not found: {img_id}")

plt.show()
```

```python

from sklearn.preprocessing import LabelEncoder
labelEncoder = LabelEncoder()
labelEncoder.fit(df['dx'])
df['label']=labelEncoder.transform(df['dx'])
```

```python
diagnoses = labelEncoder.classes_
encoded_labels = labelEncoder.transform(diagnoses)
diagnosis_mapping = dict(zip(encoded_labels, diagnoses))
for label, diagnosis in diagnosis_mapping.items():
    print(f"Label {label} represents diagnosis: {diagnosis}")
```

```python
import os
import pandas as pd

# Paths to image folders
folder1_path = 'HAM10000/HAM10000_images_part_1'
folder2_path = 'HAM10000/HAM10000_images_part_2'



# Create mapping between image_id and full image path
image_mapping = {}

# Iterate over images in folder 1
for filename in os.listdir(folder1_path):
    image_id = filename.split('.')[0]
    image_mapping[image_id] = os.path.join(folder1_path, filename)

# Iterate over images in folder 2
for filename in os.listdir(folder2_path):
    image_id = filename.split('.')[0]
    image_mapping[image_id] = os.path.join(folder2_path, filename)

# Link image paths with metadata
df['image_path'] = df['image_id'].map(image_mapping)

# Show first 5 rows
df.head()
```

```python
# Mapping from dx (short codes) to full lesion names
lesion_type_dict = {
    'akiec': 'Actinic keratoses and intraepithelial carcinoma',
    'bcc': 'Basal cell carcinoma',
    'bkl': 'Benign keratosis-like lesions',
    'df': 'Dermatofibroma',
    'mel': 'Melanoma',
    'nv': 'Melanocytic nevi',
    'vasc': 'Vascular lesions'
}

# Create new column with full lesion names
df['lesion_type'] = df['dx'].map(lesion_type_dict)

# Show updated DataFrame
df.head()
```
